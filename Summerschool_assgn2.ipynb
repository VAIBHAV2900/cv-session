# Task 1: Edge Detection using Difference of Gaussians

```python
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
einstein_img = cv2.imread('media/einstein.jpg',0)
```

```python
def Gaussian_Edge_Detector(img,std1x,std1y,std2x,std2y):
  gauss_1 = cv2.GaussianBlur(img,(5,5),std1x,std1y)
  gauss_2 = cv2.GaussianBlur(img,(9,9),std2x,std2y)

  gauss_diff = gauss_1 - gauss_2
  rev_gauss_diff = gauss_2 - gauss_1

  #Plotting the images

  plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img,cmap='gray')
  plt.figure(),plt.axis("off"),plt.title("Blurred Image by Gauss_1"),plt.imshow(gauss_1,cmap='gray')
  plt.figure(),plt.axis("off"),plt.title("Blurred Image by Gauss_2"),plt.imshow(gauss_2,cmap='gray')
  plt.figure(),plt.axis("off"),plt.title("Gauss_1 - Gauss_2"),plt.imshow(gauss_diff,cmap='gray')
  plt.figure(),plt.axis("off"),plt.title("Gauss_2 - Gauss_1"),plt.imshow(rev_gauss_diff,cmap='gray')  
  plt.show()
```

```python
Gaussian_Edge_Detector(einstein_img,0,0,150,150)
```

```python
gauss_1 = cv2.GaussianBlur(einstein_img,(5,5),0.6)
gauss_2 = cv2.GaussianBlur(einstein_img,(9,9),5)

gauss_diff=cv2.absdiff(gauss_1,gauss_2)
rev_gauss_diff = gauss_2 - gauss_1
#Plotting the images

# plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(einstein_img,cmap='gray')
# plt.figure(),plt.axis("off"),plt.title("Blurred Image by Gauss_1"),plt.imshow(gauss_1,cmap='gray')
# plt.figure(),plt.axis("off"),plt.title("Blurred Image by Gauss_2"),plt.imshow(gauss_2,cmap='gray')
plt.figure(),plt.axis("off"),plt.title("Gauss_1 - Gauss_2"),plt.imshow(gauss_diff,cmap='gray')
plt.figure(),plt.axis("off"),plt.title("Gauss_2 - Gauss_1"),plt.imshow(rev_gauss_diff,cmap='gray')
plt.show()
```
